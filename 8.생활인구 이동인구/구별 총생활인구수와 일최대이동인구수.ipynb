```python
import pandas as pd

# 예제 데이터프레임 생성
data = {'날짜': ['2023-08-18', '2023-08-19', '2023-08-20']}
df = pd.DataFrame(data)

# '년도', '월', '일' 열 추가
df['년도'] = pd.to_datetime(df['날짜']).dt.year
df['월'] = pd.to_datetime(df['날짜']).dt.month
df['일'] = pd.to_datetime(df['날짜']).dt.day

print(df)
```

```python
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
# 그래프 그리기
fig, ax = plt.subplots()

# 바그래프 그리기
bar_color = (0.2, 0.4, 0.6)
ax.bar(df['구명'], df['총생활인구수'], color=bar_color, label='총생활인구수')

# 선그래프 그리기
line_color = (0.9, 0.7, 0.1)
ax.plot(df['구명'], df['일최대이동인구수'], marker='o', color=line_color, label='일최대이동인구수')

# 라벨과 제목 설정
ax.set_ylabel('인구수')
ax.set_xlabel('구명')
plt.title('구별 총생활인구수와 일최대이동인구수')

# 범례 표시
plt.legend()

# 그래프 표시
plt.tight_layout()
plt.show()
```
